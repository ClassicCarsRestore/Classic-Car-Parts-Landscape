#### Group entries based on 'Category' , 'Supplier_type' and 'Version columns, concate in 'Title' to remove duplicated data

```python
# made a copy
db_clean2 = pd.DataFrame(db_clean)
```

```python
db_clean2.head()
```

```python
db_clean3 = db_clean2.copy()

# fill NaN in 'Version'column with 'NaN' string
db_clean3['Version'].fillna('NaN', inplace=True)

# group 'Brand', 'Supplier_type' and 'Version' and concatenate 'Supplier' to remove duplicated data
grouped_db_clean3 = db_clean3.groupby(['Brand', 'Supplier_type', 'Version']).apply(lambda x: pd.Series({
    'Title_concatenated': ', '.join(set(x['Supplier'])),
    'City': x['City'].iloc[0],
    'Country': x['Country'].iloc[0],
    'Address': x['Address'].iloc[0]
})).reset_index()

print(grouped_db_clean3.head())
```

#### Without groupby

```python

column_names = {
    'Brand': 'Category',
    'Title': 'Supplier'

}

db_clean2.rename(columns=column_names, inplace=True)

db_clean2.head()
```
